```python
import requests
import csv
import re
import urllib3

# Disable SSL warnings (use only for testing)
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

base_url = "https://future_plus.virtualmeeting.co.in/"
registration_endpoint = base_url + "index.php/welcome/save_user"

def validate_data(user_data):
    name = user_data['name'].strip()
    email = user_data['email'].strip()
    phone_no = user_data['phone'].strip()
    state = user_data['state'].strip()
    country = user_data['country'].strip()
    exprience = user_data['years exp'].strip()

    text_regex = re.compile(r'[A-Za-z -()+]+$')
    email_regex = re.compile(r'^([A-Za-z0-9_\-\.])+\@([A-Za-z0-9_\-\.])+\.([A-Za-z]{2,4})$')
    int_regex = re.compile(r'[0-9 -()+]+$')

    if len(name) < 3 or not text_regex.match(name):
        return False, "Invalid name"
    if not email_regex.match(email):
        return False, "Invalid email"
    if len(phone_no) < 10 or not int_regex.match(phone_no):
        return False, "Invalid phone number"
    if not state:
        return False, "State is required"
    if not country:
        return False, "Country is required"
    if not exprience:
        return False, "Experience is required"

    return True, "Valid"

def register_user(user_data):
    is_valid, message = validate_data(user_data)
    if not is_valid:
        print(f"Validation failed for {user_data['name']}: {message}")
        return

    form_data = {
        "name": user_data['name'].strip(),
        "email": user_data['email'].strip(),
        "phone_no": user_data['phone'].strip(),
        "city": user_data['city'].strip(),
        "state": user_data['state'].strip(),
        "country": user_data['country'].strip(),
        "special": user_data['speciality'].strip(),
        "exprience": user_data['years exp'].strip()
    }

    try:
        response = requests.post(registration_endpoint, data=form_data, verify=False)
        if response.text == "true":
            print(response.text)
            print(f"Successfully registered {user_data['name']}")
            print(f"Redirecting to: {base_url}index.php/welcome/lobby")
        else:
            print(f"Failed to register {user_data['name']}: Access Denied")
    except requests.exceptions.RequestException as e:
        print(f"Error registering {user_data['name']}: {e}")

# Read and process the CSV file
with open('users.csv', 'r') as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        register_user(row)
```

```python
import requests
import re
import urllib3

# Disable SSL warnings (use only for testing)
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

base_url = "https://future_plus.virtualmeeting.co.in/"
registration_endpoint = base_url + "index.php/welcome/save_user"

def validate_data(user_data):
    name = user_data['name'].strip()
    email = user_data['email'].strip()
    phone_no = user_data['phone'].strip()
    state = user_data['state'].strip()
    country = user_data['country'].strip()
    exprience = str(user_data['years_exp']).strip()

    text_regex = re.compile(r'[A-Za-z -()+]+$')
    email_regex = re.compile(r'^([A-Za-z0-9_\-\.])+\@([A-Za-z0-9_\-\.])+\.([A-Za-z]{2,4})$')
    int_regex = re.compile(r'[0-9 -()+]+$')

    if len(name) < 3 or not text_regex.match(name):
        return False, "Invalid name"
    if not email_regex.match(email):
        return False, "Invalid email"
    if len(phone_no) < 10 or not int_regex.match(phone_no):
        return False, "Invalid phone number"
    if not state:
        return False, "State is required"
    if not country:
        return False, "Country is required"
    if not exprience:
        return False, "Experience is required"

    return True, "Valid"

def register_user(email, user_data):
    is_valid, message = validate_data(user_data)
    if not is_valid:
        print(f"Validation failed for {user_data['name']}: {message}")
        return

    form_data = {
        "name": user_data['name'].strip(),
        "email": email.strip(),
        "phone_no": user_data['phone'].strip(),
        "city": user_data['city'].strip(),
        "state": user_data['state'].strip(),
        "country": user_data['country'].strip(),
        "special": user_data['speciality'].strip(),
        "exprience": str(user_data['years_exp']).strip()
    }

    try:
        response = requests.post(registration_endpoint, data=form_data, verify=False)
        if response.text == "true":
            print(response.text)
            print(f"Successfully registered {user_data['name']}")
            print(f"Redirecting to: {base_url}index.php/welcome/lobby")
        else:
            print(f"Failed to register {user_data['name']}: Access Denied")
    except requests.exceptions.RequestException as e:
        print(f"Error registering {user_data['name']}: {e}")

# The dictionary of users


# Process the dictionary
for email, user_data in users_dict.items():
    register_user(email, user_data)
```
